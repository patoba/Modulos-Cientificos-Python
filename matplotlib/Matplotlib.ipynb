```python
# Utilizamos este comando para especificar a matplotlib que sera usado en jupyter 
% matplotlib notebook 

import matplotlib.pyplot as plt
import numpy as np
```

### Plot

La función plot , retorna una lista de objetos de linea. Recibe como parametros arreglos de NumPy.

Se utiliza en su mayoria para realizar gráficas simples.

```python
# Variable independiente
x = np.arange(0, 10, 0.2)

# Variable dependiente
y = np.sin(x)

# Creando la figura para tener donde graficar
fig = plt.figure()

fig.suptitle("Grafica") # Agrega titulo a la figura

plt.grid() # La funcion grid cuadricula el axes sobre el figure 

# Asignando los valores para graficar en plot
fig = plt.plot(x, y) 

# Asignamos nombres a los ejes
plt.xlabel('x')
plt.ylabel('Sen(x)')

# Desplazara lo contenido que esta almacenado en Figure
plt.show()
```

```python
# Variable dependiente
y = np.sin(x)

# Creando la figura para tener donde graficar
fig = plt.figure()

fig.suptitle("Diagrama de dispersion") # Agrega titulo a la figura

plt.grid() # La funcion grid cuadricula el axes sobre el figure 

# Asignando los valores para graficar en scatter, que nos imprime cada punto
fig = plt.scatter(x, y) 

# Asignamos nombres a los ejes
plt.xlabel('x') 
plt.ylabel('Sen(x)')

# Desplazara lo contenido que esta almacenado en Figure
plt.show()
```

Existe otra forma de graficar con mas metodos, esta es utlizando axes.  Axes es la clase que hace referencia a todo lo relacionado con la grafica.

### Axes
Es la 'gráfica'.Es la región donde esta representado el espacio de datos. Un figur puede contener varios Axes pero un Axes solo puede ser contenido en un objeto Figure.

Axes cuenta con los metodos, plot, scatter y subplot. Utilizados para la graficación de datos

### Axis
Es la clase encargada de definir los limites de la grafica y generar la métrica (las marcas en cada eje), además de las etiquetas en cada marca. 


Todas las funciones para gráficar esperan recibir como parametros arreglos de NumPy.

La primera llamada a la función plt.plot() creara el Figure necesario para poder agregar Axes. Una vez creado uno , siempre se reutilizara la figura que se creo por primera vez. En jupyter creará una figure por cada celda.



```python
# Variable independiente
x = np.arange(0, 10, 0.2)

# Variable dependiente
y = np.sin(x)

# Creamos la figura
fig = plt.figure()

# Creamo un axes para tener en la figura
ax = plt.axes()
ax.set_xlabel("x")
ax.set_ylabel("Sen(x)")
ax.set_title("Hola ")
ax.grid()


# Utilizamos el metodo plot parra 
line = ax.plot(x, y)
```

### Subplots

Si queremos visualizar mas de una grafica en una misma figura podemos utilizar subplots.

La función subplots retorna un Figure y Axes. 

Subplots recibe como parametros el numero de filas y columnas.

plt.subplots(filas, columnas)

Támbien podemos definir si comparten un eje.

Con el parametro sharey(Compartir y) o sharex.

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)

#Graficar solo una vez
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title('Simple plot')
```

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)


#Crea dos axes 
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True) # Retorna los Axes como tuplas
ax1.plot(x, y)
ax1.set_title('Compartiendo eje Y')
ax2.scatter(x, y)
```

```python
#First create some toy data:
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)


#Crer dos axes de manera vertical
fig, (ax3, ax4) = plt.subplots(2, 1, sharex=True) # 2 filas y una columna
ax3.plot(x, y)
ax3.set_title('Compartiendo Eje X')
ax4.scatter(x, y)
```
